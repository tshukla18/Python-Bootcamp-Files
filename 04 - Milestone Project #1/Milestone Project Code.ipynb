```python
from IPython.display import clear_output

def display_board(board):
    print('\n')   
    print(f"   |   |   \n {board[6]} | {board[7]} | {board[8]}  \n   |   |   ")
    print(f"-----------")
    print(f"   |   |   \n {board[3]} | {board[4]} | {board[5]} \n   |   |   ")
    print(f"-----------")
    print(f"   |   |   \n {board[0]} | {board[1]} | {board[2]}  \n   |   |   ")
```

```python
#2) Input by Player 1: while loop continues asking until player gives correct answer
```

```python
def player_input():
    global player1_marker
    global player2_marker
    global readyOrNot
    check1 = True
    check2 = True
    
    while check2:
        while check1:
            player1_marker = input("\nPlayer 1: Do you want to be X or O?: ").upper()
        
            if player1_marker == 'X':
                player2_marker = 'O' 
                check1 = False
            elif player1_marker == 'O':
                player2_marker = 'X' 
                check1 = False
                print("Player 1 will go first.\n")
                
            #print("Player {randonNum} will go first.\n")
                
        readyOrNot = input("Are you ready to play? Enter Yes or No. ")
                
        if readyOrNot == 'Yes' or readyOrNot == 'No':
            check2 = False
```

```python
#3) Board list object, a marker ('X' or 'O'), and a desired position (number 1-9) are paramenters
```

```python
def place_marker(board, marker, position):
    if marker == 'X' or marker == 'O':
        board[int(position)-1] = marker
```

```python
#4) Takes in a board and a marker (X or O) to check if that marker has won
```

```python
def win_check(board, mark):
    
    if mark == 'X' or mark == 'O':
        
        return((board[0] == board[1] == board[2] == mark) or ##### ROWS #####
        (board[3] == board[4] == board[5] == mark) or 
        (board[6] == board[7] == board[8] == mark) or
        (board[0] == board[3] == board[6] == mark) or ##### COLUMNS #####
        (board[1] == board[4] == board[7] == mark) or
        (board[2] == board[5] == board[8] == mark) or
        (board[0] == board[4] == board[8] == mark) or ##### DIAGONALS #####
        (board[2] == board[4] == board[6] == mark))
```

```python
#6) Check if space on board is available
```

```python
def space_check(board, position):
    if board[int(position)-1] == ' ':
        return True
    else:
        return False
```

```python
#7) Checks whether the board is full.Returns a boolean value. 
```

```python
def full_board_check(board):
    
    for value in board:
        if value == ' ':
            return False
        
    return True
```

```python
#8) Asks for a player's next position (1-9)
#uses fn from step 6 to check if it's a free position. If it is, return the position for later use.
```

```python
def player_choice(board):
    global position 
    position = input("What position do you want? ")
    
    if space_check(board, position):
        return position
    else:
        return False
```

```python
##### WELCOME #####
print('Welcome to Tic Tac Toe!\n')
print('Game positions 1-9:')
position_board = ['1','2','3','4','5','6','7','8','9']
display_board(position_board)
counter = True

while counter:
    ##### SET-UP #####
    game_on = True

    turn = choose_first()
    player_input()

    if readyOrNot == 'No':
        print('\nThanks for playing!')
        break

    ##### GAMEPLAY #####
    game_board = [' ',' ',' ',' ',' ',' ',' ',' ',' ']
    display_board(game_board)

    ##### PLAYER 1 TURN #####
    while game_on:
        if turn == 'Player 1':
            print('Player 1, your turn')
            player_choice(game_board)
            place_marker(game_board, player1_marker, position)
            display_board(game_board)

            if win_check(game_board, player1_marker):
                print('~~ Congratulations! Player 1 has won the game!! ~~')
                game_on = False
            else:
                if full_board_check(game_board):
                    print('~~ The game is a draw! ~~')
                    game_on = False
                else:
                    turn = 'Player 2'
        else:
        ##### PLAYER 2 TURN #####
            print('Player 2, your turn')
            player_choice(game_board)
            place_marker(game_board, player2_marker, position)
            display_board(game_board)

            if win_check(game_board, player2_marker):
                print('~~ Congratulations! Player 2 has won the game!! ~~')
                game_on = False
            else:
                if full_board_check(game_board):
                    print('~~ The game is a draw! ~~')
                    break
                else:
                    turn = 'Player 1'

    ##### END GAME OR PLAY ANOTHER #####
    print('\n##### GAME OVER #####')
    
    if not replay():
        counter = False
```
